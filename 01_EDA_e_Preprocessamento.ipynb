## 1: EDA e Pré-processamento

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import locale
import numpy as np

# ========== LOCALE PARA BRASIL ==========
try:
    locale.setlocale(locale.LC_ALL, 'pt_BR.UTF-8')
except:
    try:
        locale.setlocale(locale.LC_ALL, 'Portuguese_Brazil.1252')
    except:
        print("Locale pt_BR não encontrado. Os números serão impressos no formato padrão.")

os.makedirs('img', exist_ok=True)
os.makedirs('export', exist_ok=True)

# 1. LEITURA DOS DADOS
df = pd.read_csv('data/UCI_Credit_Card.csv')

# 2. RENOMEANDO COLUNAS E MAPEANDO
rename_dict = {
    'ID': 'id',
    'LIMIT_BAL': 'limite_credito',
    'SEX': 'sexo',
    'EDUCATION': 'escolaridade',
    'MARRIAGE': 'casamento',
    'AGE': 'idade',
    'PAY_0': 'atraso_0',
    'PAY_2': 'atraso_2',
    'PAY_3': 'atraso_3',
    'PAY_4': 'atraso_4',
    'PAY_5': 'atraso_5',
    'PAY_6': 'atraso_6',
    'BILL_AMT1': 'fatura_mes1',
    'BILL_AMT2': 'fatura_mes2',
    'BILL_AMT3': 'fatura_mes3',
    'BILL_AMT4': 'fatura_mes4',
    'BILL_AMT5': 'fatura_mes5',
    'BILL_AMT6': 'fatura_mes6',
    'PAY_AMT1': 'pagto_mes1',
    'PAY_AMT2': 'pagto_mes2',
    'PAY_AMT3': 'pagto_mes3',
    'PAY_AMT4': 'pagto_mes4',
    'PAY_AMT5': 'pagto_mes5',
    'PAY_AMT6': 'pagto_mes6',
    'default.payment.next.month': 'inadimplente'
}
df.rename(columns=rename_dict, inplace=True)
sexo_map = {1: 'Masculino', 2: 'Feminino'}
escolaridade_map = {1: 'Pós-graduação', 2: 'Universidade', 3: 'Ensino médio', 4: 'Outros', 5: 'Desconhecido', 6: 'Desconhecido', 0: 'Desconhecido'}
casamento_map = {1: 'Casado', 2: 'Solteiro', 3: 'Outros', 0: 'Desconhecido'}
df['sexo'] = df['sexo'].map(sexo_map)
df['escolaridade'] = df['escolaridade'].map(escolaridade_map)
df['casamento'] = df['casamento'].map(casamento_map)
df['inadimplente'] = df['inadimplente'].map({1: 'Sim', 0: 'Não'})

# 3. FEATURE ENGINEERING - Proporção paga
for i in range(1, 7):
    pay_amt_col = f'pagto_mes{i}'
    bill_amt_col = f'fatura_mes{i}'
    if pay_amt_col in df.columns and bill_amt_col in df.columns:
        df[f'proporcao_pago_{i}'] = df[pay_amt_col] / (df[bill_amt_col] + 1)

# 3.1 Tratamento de inf e -inf gerados
df.replace([np.inf, -np.inf], np.nan, inplace=True)

# 4. DADOS RESUMIDOS
print("Primeiras linhas do DataFrame:")
display(df.head())
print("\nInfo do DataFrame:")
print(df.info())
print("\nEstatísticas descritivas:")
display(df.describe(include='all').T)
print("\nValores únicos por coluna:")
print(df.nunique())
print("\nDistribuição do target:")
print(df['inadimplente'].value_counts(normalize=True).map('{:.1%}'.format))
print("\nDistribuição sexo:")
print(df['sexo'].value_counts())
print("\nDistribuição escolaridade:")
print(df['escolaridade'].value_counts())
print("\nDistribuição casamento:")
print(df['casamento'].value_counts())

# 5. FUNÇÃO DE GRÁFICO PADRÃO
def plot_bar_with_labels(labels, values, title, xlabel, imgname, rotation=0):
    plt.figure(figsize=(8,4))
    cor_roxo = "#6927c9"
    bars = plt.bar(labels, values, color=cor_roxo, edgecolor='black')
    plt.title(title, fontsize=16, pad=15)
    plt.xlabel(xlabel, fontsize=12)
    plt.ylabel('Quantidade', fontsize=12)
    plt.xticks(rotation=rotation, fontsize=11)
    plt.yticks(fontsize=12)
    plt.gca().set_facecolor("white")
    plt.grid(axis='y', linestyle='--', alpha=0.12)
    max_val = max(values)
    plt.ylim(0, max_val * 1.30)
    for bar, val in zip(bars, values):
        perc = val / sum(values) * 100
        plt.text(
            bar.get_x() + bar.get_width()/2, val + (max_val*0.05),
            f"{locale.format_string('%d', val, grouping=True)}\n({perc:.1f}%)",
            ha='center', va='bottom',
            fontsize=12, fontweight='bold', color="black", linespacing=1.2
        )
    plt.tight_layout()
    plt.savefig(f'img/{imgname}.png', dpi=180, transparent=False)
    plt.show()

# 6. PLOTS EXPLORATÓRIOS
# Distribuição do target
contagem = df['inadimplente'].value_counts()
plot_bar_with_labels(
    contagem.index, contagem.values,
    'Distribuição da Inadimplência',
    'Inadimplente', 'distribuicao_inadimplente'
)
# Sexo
sexo_cont = df['sexo'].value_counts()
plot_bar_with_labels(
    sexo_cont.index, sexo_cont.values,
    'Distribuição por Sexo',
    'Sexo', 'distribuicao_sexo'
)
# Escolaridade
esc_cont = df['escolaridade'].value_counts()
plot_bar_with_labels(
    esc_cont.index, esc_cont.values,
    'Distribuição por Escolaridade',
    'Escolaridade', 'distribuicao_escolaridade', rotation=0
)
# Estado Civil
cas_cont = df['casamento'].value_counts()
plot_bar_with_labels(
    cas_cont.index, cas_cont.values,
    'Distribuição por Estado Civil',
    'Estado Civil', 'distribuicao_estado_civil', rotation=0
)

# 7. ANÁLISE DE NULOS FINAL
print('\nNulos após feature engineering:')
print(df.isnull().sum().sort_values(ascending=False).head(10))

# 8. EXPORTAÇÃO DA BASE FINAL
df.to_csv('export/dataset_preparado_para_modelo.csv', index=False)
print('\nArquivo export/dataset_preparado_para_modelo.csv salvo para o próximo notebook!')
```
